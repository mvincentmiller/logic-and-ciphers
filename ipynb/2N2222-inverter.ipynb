```python
import SchemDraw as schem
import SchemDraw.elements as e
d = schem.Drawing(unit=2.5)

Q1 = d.add(e.BJT_NPN_C, label="")
d.add(e.LINE, xy=Q1.collector, d='up', l=d.unit/2)
D1 = d.add(e.DOT)
d.add(e.LINE, d='right', l=2)
d.add(e.LINE, d='down', l=1)
LED = d.add(e.LED, d='down', toy=D2.start)
d.add(e.LINE, xy=Q1.base, d='left', l=d.unit/2)
BUTTON = d.add(e.BUTTON, d='up')
R1 = d.add(e.RES, d='up', label='$1k$')
d.add(e.DOT)
d.add(e.LINE, d='right', l=2)
d.add(e.DOT)
R2 = d.add(e.RES, d='down', toy=D1.start, label="100Ω")
d.add(e.LINE, xy=Q1.emitter, d='down', l=d.unit/2)
D2 = d.add(e.DOT)
d.push()
d.add(e.LINE, d='left', l=4)
d.add(e.LINE, d='up')
Vt = d.add(e.BATTERY, d='up', label='$5v$', reversed=True)
V = d.add(e.LINE, d='up', l=1.937)
d.labelI(V, arrowlen=1.5, arrowofst=0.5)
d.add(e.LINE, d='right', l=2)
d.pop()
d.add(e.LINE, d='right', l=2)

d.draw()
# d.save('inverter.png')
```
